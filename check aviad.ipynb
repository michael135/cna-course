```python

def read_all(train_f, test_f, graph_f):
    with open(train_f) as file:
        train = eval(file.read())
    with open(test_f) as file:
        test = eval(file.read())

    G = nx.Graph()
    with open(graph_f) as file:
        content = file.readlines()
    for line in content:
        edge = list(eval(line))
        G.add_edge(edge[0], edge[1], weight=edge[2])

    return train, test, G


```
